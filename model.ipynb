```python
import pdfplumber

def extract_text_from_pdf(file_path: str) -> str:
    text = ""

    with pdfplumber.open(file_path) as pdf:
        for page in pdf.pages:
            text += page.extract_text() or ""

    return text.strip()
```

```python
prompt = f'''
You are an experienced AI/ML Technical Interviewer conducting a mock interview for a candidate.

Your objective is to simulate a realistic interview and assess the candidate's:

* Technical Depth
* Communication Skills
* Confidence
* Problem Solving Ability
* Project Understanding

You have access to:

1. Candidate Resume
2. Job Description
3. Previous Interview Transcript

Interview Guidelines:

1. Ask ONLY ONE question at a time.

2. Output ONLY the next interview question.
   Do not provide explanations, analysis, feedback, commentary, observations, introductions, or reasoning.

3. Conduct the interview naturally like a real human interviewer.

4. Progress through the interview in stages:

Stage 1 - Introduction

* Background
* Education
* Interests
* Career goals
* Motivation

Stage 2 - Project Discussion

* Project overview
* Project architecture
* Design decisions
* Challenges faced
* Learnings

Stage 3 - Technical Evaluation

* AI/ML concepts
* Machine Learning
* Deep Learning
* Generative AI
* RAG
* LLMs
* Vector Databases
* LangChain
* Python

Stage 4 - Behavioral Questions

* Teamwork
* Problem solving
* Leadership
* Communication
* Learning ability

5. Questions should gradually increase in difficulty.

6. Do not abruptly jump from introduction questions to highly advanced technical questions.

7. Build questions using:

* Candidate's previous answers
* Candidate's resume
* Job description

8. Avoid repeating questions that have already been asked.

9. Follow up when the candidate provides useful information.

Example:
Candidate:
"I built a Resume Screening System."

Good Follow-up:
"Can you walk me through the architecture of that system?"

Bad Follow-up:
"What is overfitting in machine learning?"

Do NOT repeatedly ask advanced technical questions when the candidate is not engaging.

11. Use answer quality to control depth.

If the candidate gives:

* detailed answers
* technical explanations
* project examples

Then:

* ask deeper follow-up questions
* challenge assumptions
* explore implementation details

12. Prioritize projects mentioned in the resume before asking theoretical questions.

13. If the candidate demonstrates strong knowledge in a topic, explore that topic further.

14. Keep questions concise.
    Maximum 2 sentences.

15. Never mention:

* transcript analysis
* interview stages
* candidate engagement score
* internal reasoning
* evaluation criteria

16. 
The candidate's MOST RECENT answer should be your primary source for generating the next question.

Priority order:

1. Candidate's latest answer
2. Previous interview history
3. Resume
4. Job description

Whenever possible, ask a follow-up question based on the candidate's latest answer before introducing a new topic.

Return ONLY the next interview question.



 
resume_text = {text}
job description = {jd}
conversation = {conversation}
'''
```

```python
question_weights = {
        "introduction": {
            "confidence": 0.9,
            "communication": 0.8,
            "technical_depth": 0.1
        },
        "technical": {
            "confidence": 0.3,
            "communication": 0.4,
            "technical_depth": 1.0
        },
        "behavioral": {
            "confidence": 0.8,
            "communication": 0.9,
            "technical_depth": 0.2
        }
    }
```

```python
import sounddevice as sd
from scipy.io.wavfile import write
from groq import Groq

client = Groq(api_key=api_key)

def speech_to_text(duration=15):
    
    fs = 16000

    print("Speak now...")

    audio = sd.rec(
        int(duration * fs),
        samplerate=fs,
        channels=1,
        dtype="int16"
    )

    sd.wait()

    write("answer.wav", fs, audio)

    with open("answer.wav", "rb") as file:
        transcription = client.audio.transcriptions.create(
            file=file,
            model="whisper-large-v3"
        )

    print("You said:", transcription.text)

    return transcription.text
```

```python
def interview(model, prompt,question_weights,limit=5, voice_mode=False):
    count = 0
    conversation = []

    question = "Tell me about yourself"

    interview_history = ""
    

    while True:
        if count >= limit:
            break
        
        print("\nInterviewer:", question)
        if voice_mode:
            user_answer = speech_to_text()
        else:
            user_answer = input("You: ")
            print("\nUser:",user_answer)

        if user_answer.lower() == "end":
            break

        full_prompt = f"""
        {prompt}

        Here is the running transcript of the interview so far.
        {interview_history}

        
         """

        response = model.invoke(full_prompt)
        classification_prompt = f'''
    You are an interview question classifier.

    Your task is to classify the interview question into EXACTLY ONE of the following categories:

    Categories:
    - introduction
    - technical
    - behavioral

    Definitions:

    introduction:
    Questions about self-introduction, background, education, interests, career goals, strengths, weaknesses, and personal profile.

    technical:
    Questions testing technical concepts, algorithms, machine learning, deep learning, RAG, LLMs, databases, programming, system design, or theoretical knowledge.

    behavioral:
    Questions about teamwork, leadership, conflict resolution, communication, decision-making, problem-solving situations, failures, achievements, and work experiences.

    Rules:
    - Return ONLY the category name.
    - Return exactly one category.
    - Do not explain your reasoning.
    - Do not output any additional text.
    - Do not output punctuation.

    Question:
    {question}


'''

        question_type = model.invoke(classification_prompt).content.lower()

        conversation.append({
            "question": question,
            "answer": user_answer,
            "question_type": question_type,
            "weights": question_weights[question_type],
            "answer_length": len(user_answer.split())
        })

        

        for qa in conversation:
            interview_history += f"""
            Interviewer: {qa['question']}
            Candidate: {qa['answer']}
            """

        question = response.content

        count += 1

    return conversation
```

```python
transcript = interview(model, prompt,question_weights,5,False)
```

```python
def retrieve_by_aspect(transcript, aspect, threshold):
    return [
        qa
        for qa in transcript
        if qa["weights"][aspect] >= threshold
    ]
```

```python
confidence_qas = transcript

technical_qas = retrieve_by_aspect(
    transcript,
    aspect="technical_depth",
    threshold=0.8
)

communication_qas = transcript
```

```python
def retrieval_evaluation (confidence_prompt, technical_depth_prompt, communication_prompt, model):
    confidence_report = model.invoke(confidence_prompt)

    technical_depth_report = model.invoke(technical_depth_prompt)

    communication_report = model.invoke(communication_prompt)

    return confidence_report, technical_depth_report, communication_report
```

```python
communication_prompt = f'''
You are an experienced interviewer.

Evaluate ONLY the candidate's communication skills based on the give questions and answers:
{communication_qas}

Focus on:
- Clarity of explanations
- Organization of thoughts
- Completeness of answers
- Professional communication style
- Ability to express ideas effectively

Provide:
1. Communication score (1-10)
2. Evidence from the transcript
3. Strengths
4. Areas for improvement'''
```

```python
confidence_report, technical_depth_report, communication_report = retrieval_evaluation(confidence_prompt, technical_depth_prompt, communication_prompt, model)
```
